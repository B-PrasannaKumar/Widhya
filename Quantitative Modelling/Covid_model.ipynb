```python
import pandas as pd
```

```python
covid_data = pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv')
```

```python
covid_data
```

```python
date_data = covid_data.groupby(['Date'], sort=False).sum()
```

```python
date_data.head()
```

```python
date_data['totals'] = date_data['ConfirmedForeignNational'] + date_data['ConfirmedIndianNational'] + date_data['Cured'] + date_data['Deaths']
```

```python
date_data
```

```python
answer = int(date_data.groupby('Date').get_group('04/03/20')['totals'])
```

```python
import matplotlib.pyplot as plt
```

```python
date_data.columns
```

```python
x = date_data.reset_index()['Date']
y = date_data['totals']
```

```python
plt.figure(figsize = (15,5))
plt.xlabel('Dates')
plt.ylabel('Total Cases')
plt.xticks(rotation = 90)
plt.plot(x,y)
```
