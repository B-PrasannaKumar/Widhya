# Task 2

```python
import pandas as pd
```

```python
insta_data = pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/instagram_reach.csv')
```

```python
insta_data
```

```python
insta_data['Time since posted'] = [int(i.replace('hours', '')) for i in list(insta_data['Time since posted'])]
```

```python
insta_data
```

# Task 3

```python
from sklearn.model_selection import train_test_split
```

```python
x = insta_data[['Followers', 'Time since posted']]
y = insta_data['Likes']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
result = lr.predict(x_test)
```

```python
import pickle
```

```python
filename = 'widhya_ml.sav'
pickle.dump(lr, open(filename, 'wb'))
```
